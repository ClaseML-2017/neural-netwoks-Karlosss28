## Ejercicio 1.a: AND con regresión logística

#### Datos de entrenamiento

```python
X_ej_1a=np.asarray([[0,0],[0,1],[1,0],[1,1]])
Y_ej_1a=np.asarray([[0],[0],[0],[1]])
```

#### Variables / Placeholders

```python
iterations_ej_1a = 5000
cte_aprendizaje_ej_1a = 0.01
input_size_ej_1a = 2
output_layer_size_ej_1a = 1

x_ej_1a = tf.placeholder(tf.float32, [None, input_size_ej_1a])
y_ej_1a_ = tf.placeholder(tf.float32, [None, output_layer_size_ej_1a]) 

W_layer1_ej_1a = tf.Variable(tf.random_uniform([input_size_ej_1a, output_layer_size_ej_1a], -1, 1), name="W_layer1_ej_1a")
b_layer1_ej_1a = tf.Variable(tf.zeros([output_layer_size_ej_1a]), name="b_layer1_ej_1a")
```

#### Grafo

```python
y_ej_1a = tf.nn.sigmoid(tf.add(tf.matmul(x_ej_1a,W_layer1_ej_1a),b_layer1_ej_1a))
lossfn_ej_1a = tf.reduce_mean(tf.reduce_sum((y_ej_1a_-y_ej_1a)**2))
train_step_ej_1a = tf.train.GradientDescentOptimizer(cte_aprendizaje_ej_1a).minimize(lossfn_ej_1a)
```

#### Inicialización de la sesión

```python
init = tf.global_variables_initializer()
sess = tf.Session()
```

#### Correr Programa

```python
sess.run(init)
for i in range(iterations_ej_1a):
  sess.run(train_step_ej_1a, feed_dict={x_ej_1a: X_ej_1a, y_ej_1a_: Y_ej_1a})
```

#### Accuracy

```python
correct_prediction_ej_1a = tf.equal(tf.round(y_ej_1a),y_ej_1a_)
accuracy_ej_1a = tf.reduce_mean(tf.cast(correct_prediction_ej_1a, tf.float32))
print(sess.run(accuracy_ej_1a, feed_dict={x_ej_1a: X_ej_1a, y_ej_1a_: Y_ej_1a}))
```

### Visualización

#### Barrera de Decisión

```python
w_ej_1a=[sess.run(b_layer1_ej_1a, feed_dict={x_ej_1a: X_ej_1a, y_ej_1a_: Y_ej_1a})[0]]+[i[0] for i in sess.run(W_layer1_ej_1a, feed_dict={x_ej_1a: X_ej_1a, y_ej_1a_: Y_ej_1a})]
m_ej_1a=((w_ej_1a[0]/w_ej_1a[2]))/((-w_ej_1a[0]/w_ej_1a[1]))
```

```python
plt.scatter(X_ej_1a[:,0],X_ej_1a[:,1],c=['red' if i==1 else 'blue' for i in Y_ej_1a])
x2_ej_1a = np.linspace(-3,4,100)
plt.plot(x2_ej_1a,-w_ej_1a[0]/w_ej_1a[2]+m_ej_1a*x2_ej_1a,color='red')
plt.ylim([-3,4])
plt.xlim([-3,4])
```

#### Cerramos la sesión

```python
sess.close()
```

### Ejercicio 3: Puntos dentro de un círculo con ANN

```python
plt.scatter(X_ej_3[:,0],X_ej_3[:,1], c=['red' if i==1 else 'blue' for i in Y_ej_3])
plt.ylim([0,1])
plt.xlim([0,1])
```

### ANN con dos capas ocultas y diferente número de neuronas en dichas capas (Función generadora)

```python
def ann_dos_capas_n_neuronas(neurona):
    neurons_per_layer_ej_3=neurona
    
    # Variables / Placeholders

    W_layer1_ej_3b = tf.Variable(tf.random_uniform([input_size_ej_3, neurons_per_layer_ej_3], -1, 1), name="W_layer1_ej_3b")
    W_layer2_ej_3b = tf.Variable(tf.random_uniform([neurons_per_layer_ej_3,neurons_per_layer_ej_3], -1, 1), name="W_layer2_ej_3b")
    W_layer3_ej_3b = tf.Variable(tf.random_uniform([neurons_per_layer_ej_3,output_layer_size_ej_3], -1, 1), name="W_layer3_ej_3b")
    b_layer1_ej_3b = tf.Variable(tf.zeros([neurons_per_layer_ej_3]), name="b_layer1_ej_3b")
    b_layer2_ej_3b = tf.Variable(tf.zeros([neurons_per_layer_ej_3]), name="b_layer2_ej_3b")
    b_layer3_ej_3b = tf.Variable(tf.zeros([output_layer_size_ej_3]), name="b_layer3_ej_3b")
    
    # Grafo
    hidden_1_ej_3b = tf.nn.sigmoid(tf.matmul(x_ej_3, W_layer1_ej_3b) + b_layer1_ej_3b)
    hidden_2_ej_3b = tf.nn.sigmoid(tf.matmul(hidden_1_ej_3b, W_layer2_ej_3b) + b_layer2_ej_3b)
    y_ej_3b = tf.nn.sigmoid(tf.matmul(hidden_2_ej_3b, W_layer3_ej_3b) + b_layer3_ej_3b)

    lossfn_ej_3b = tf.reduce_mean(tf.reduce_sum((y_ej_3_-y_ej_3b)**2)) #cuadratico
    train_step_ej_3b = tf.train.GradientDescentOptimizer(cte_aprendizaje_ej_3).minimize(lossfn_ej_3b)	

    # Inicialización de la sesión
    init = tf.global_variables_initializer()
    sess = tf.Session()
    
    # Correr el programa
    sess.run(init)
    for i in range(iterations_ej_3):
      sess.run(train_step_ej_3b, feed_dict={x_ej_3: X_ej_3, y_ej_3_: Y_ej_3})
    
    # Accuracy
    correct_prediction_ej_3b = tf.equal(tf.round(y_ej_3b),y_ej_3_)
    accuracy_ej_3b = tf.reduce_mean(tf.cast(correct_prediction_ej_3b, tf.float32))
    accuracy_ej_3b_calc = sess.run(accuracy_ej_3b, feed_dict={x_ej_3: X_ej_3, y_ej_3_: Y_ej_3})
    accuracies_2.append(accuracy_ej_3b_calc)
    
    print("\n\nNúmero de neuronas: "+str(neurona)+".\nAccuracy: "+str(accuracy_ej_3b_calc))
    
    #Visualizacion
    x_vis_ej_3b = np.random.rand(2000,2)
    y_vis_ej_3b = sess.run(y_ej_3b, feed_dict={x_ej_3:x_vis_ej_3b})
    y_vis_ej_3b = np.round(y_vis_ej_3b)
    
    plt.scatter(x_vis_ej_3b[:,0],x_vis_ej_3b[:,1], c=['red' if i==1 else 'blue' for i in y_vis_ej_3b])
    plt.ylim([0,1])
    plt.xlim([0,1])
    
    plt.show()
    plt.clf()
    sess.close
```

#### Variables generales

```python
neuronas = [3,5,7,9]

iterations_ej_3 = 50000
cte_aprendizaje_ej_3 = 0.01

input_size_ej_3=2
output_layer_size_ej_3=1

accuracies=[]
accuracies_2=[]

x_ej_3 = tf.placeholder(tf.float32, [None, input_size_ej_3])
y_ej_3_ = tf.placeholder(tf.float32, [None, output_layer_size_ej_3])
```

### Resultados con una capa oculta

```python
for neurona in neuronas:
    ann_una_capa_n_neuronas(neurona)
```
